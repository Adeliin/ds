### Задание 6 (1 балл): убрать пропущенные значения из данных. На примере признака Unit price.

Существует несколько методов работы с пропущенными данными в зависимости от количества пропусков.
Если пропусков в столбце больше 40%, следует удалять столбец. Если меньше 40%, можно либо удалить 
отдельные строчки с пропущенными значениями в столбце, либо заменить их на другое значение. В качестве заменяемого знаения могут служить: 1. Мера центральной тенденции по всему столбцу; 2. Мера центральной тенденции внутри группы; 3. Случайный элемент.

```python
#     В случае с признаком "Unit price" пропущенных значений нет (убедилась выше). Однако, если представить, что они есть, 
# то в зависимости от количества пропусков,  нужно было бы воспользоваться одним из вышеперечисленных способов. В качестве меры 
# центральной тенденции стоит подбирать показатель для непрерывных данных в зависимости от распределения
#     Найдем распределение значений столбца "Unit price".

df["Unit price"].hist(bins=40)
```

```python
# Так как распределение не очень симметричное, я бы заменила пропущенные значения в столбце на медиану, следующим образом:
df['Unit price'] = df['Unit price'].fillna(df['Unit price'].median())
```

```python
# Обращаясь к альтернативному методу борьбы с пропущенными значениями, когда пропусков меньше 40%, 
# можно было бы исключить только строки с пропущенными значениями.
df.dropna(subset = ['Unit price'])
```

```python
# Если бы пропусков было бы более 40%:
# я бы удалила столбец и сохранила бы в новую переменную. Сохранять в новую переменную я не буду, так как в данном датасете 
# пропуски отсутствуют.

df.drop(['Unit price'], axis=1) 
```

### Задание 7 (1 балл): придумать новый признак (колонку) для наших данных.

```python
# Создам новый признак, который будет интерпретироваться, как рискованность ведения бизнеса в зависимости от города.
# Для этого определю функцию, которая будет считать долю риска, котоому подвергается бизнес в зависимости от местоположения. 
# Числа были подобраны случайно
def calculate_risk(city):
    if city == 'Mandalay':
        return 0.15
    elif city == 'Yangon':
        return 0.25
    elif city == 'Naypyitaw':
        return 0.45
        
df['Risk'] = df['City'].apply(calculate_risk)
```
